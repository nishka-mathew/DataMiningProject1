### Section 2 - Data Analysis with Visualizations

```python
#Distribution of Age in the Survey
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=False, color='#69d', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Paired'))
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
#Gender vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.countplot(x="Gender", hue="Decided to Pursue MBA?", data=df, palette="Paired")
plt.title("Gender vs. MBA Decision")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Undergraduate Major Distribution
plt.figure(figsize=(8, 8))
df['Undergraduate Major'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Undergraduate Major Distribution')
plt.ylabel('')
plt.show()
```

```python
#Undergraduate Major vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.countplot(x="Undergraduate Major", hue="Decided to Pursue MBA?", data=df, palette="viridis")
plt.title("Undergraduate Major vs. MBA Decision")
plt.xlabel("Undergraduate Major")
plt.ylabel("Count")
plt.show()
```

```python
#Management Experience Distribution
plt.figure(figsize=(6, 6))
df['Has Management Experience'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Paired'))
plt.title('Management Experience Distribution')
plt.ylabel('')
plt.show()
```

```python
#Management Experience vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.countplot(x="Has Management Experience", hue="Decided to Pursue MBA?", data=df, palette="crest")
plt.title("Management Experience vs. MBA Decision")
plt.xlabel("Has Management Experience")
plt.ylabel("Count")
plt.show()
```

```python
#Reason for MBA vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.countplot(x="Reason for MBA", hue="Decided to Pursue MBA?", data=df, palette="mako")
plt.title("Reason for MBA vs. MBA Decision")
plt.xlabel("Reason for MBA")
plt.ylabel("Count")
plt.show()
```

```python
#Reported Annual Salary Before MBA Based on Years of Work Experience
plt.figure(figsize = (10,5))
sns.lineplot(x = 'Years of Work Experience', y = 'Annual Salary (Before MBA)', data = df, marker='o', color='green')
plt.xlabel('Years of Work Experience')
plt.ylabel('Annual Salary (Before MBA)')
plt.title('Reported Annual Salary Before MBA Based on Years of Work Experience')
plt.show()
```

```python
#Reported GRE/GMAT Score Based on Years of Work Experience
plt.figure(figsize=(10, 5))
sns.lineplot(x='Years of Work Experience', y='GRE/GMAT Score', data=df, marker='o') 
plt.xlabel('Years of Work Experience')
plt.ylabel('GRE/GMAT Score')
plt.title('Reported GRE/GMAT Score Based on Years of Work Experience')
plt.show()
```

```python
#Reported GRE/GMAT Score vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.histplot(df, x="GRE/GMAT Score", hue="Decided to Pursue MBA?", kde=True, bins=30, palette="mako", multiple="layer", alpha=0.6)
plt.title("Reported GRE/GMAT Score vs. MBA Decision")
plt.xlabel("GRE/GMAT Score")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#Expected Post-MBA Salary vs. MBA Decision
plt.figure(figsize=(8, 5))
sns.histplot(df, x="Expected Post-MBA Salary", hue="Decided to Pursue MBA?", kde=True, bins=30, palette="viridis", multiple="layer", alpha=0.6)
plt.title("Expected Post-MBA Salary vs. MBA Decision")
plt.xlabel("Expected Post-MBA Salary")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
